```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

### Subplots

```python
np.random.seed(42)
x = np.random.random(1000)
grocery_list = {
    'Ghee': 10,
    'Rice': 35,
    'Wheat': 20,
}
```

```python
# Generate a figure with 2x2 subplots, each containing a different type of plot (e.g., line plot, scatter plot, bar plot, and histogram).
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
    nrows=2, ncols=2, figsize=(10, 8)
)

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(grocery_list.keys(), grocery_list.values())
ax4.hist(np.random.randn(100));

# Customize the subplots with titles, axis labels, and legends.
ax1.set(
    title='Line Plot',
    xlabel='X-axis',
    ylabel='Y-axis'
)
ax1.legend(['Line 1'])

ax2.set(
    title='Scatter Plot',
    xlabel='X-axis',
    ylabel='Y-axis'
)
ax2.legend(['Scatter 1'])

ax3.set(
    title='Bar Plot',
    xlabel='Items',
    ylabel='Quantity'
)
ax3.legend(['Grocery Bar'])

ax4.set(
    title='Histogram',
    xlabel='Values',
    ylabel='Frequency',
)
ax4.legend(['Histogram 1'])

fig.tight_layout()
plt.show()
```

```python
# Create a histogram for a given numerical data set. Customize the number of bins.
np.random.seed(42)
data = np.random.randn(1000)
plt.hist(data, bins=20);
```

```python
# Plot multiple histograms on the same figure to compare different data sets.
# importing dataset
df = pd.read_csv('https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales.csv')
df['Price'] = df['Price'].replace('[\$\,\.]', '', regex=True).astype(int)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df['Price'], bins=10);
ax.hist(df['Odometer (KM)'], bins=10);
ax.legend(['Price', 'Odometer']);
```

```python
# Normalize the histograms and add appropriate labels and titles
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df['Price'], density = True, bins=10, label='Price');
ax.hist(df['Odometer (KM)'], density = True, bins=10, label='Odometer (KM)');

# Adding labels and title
ax.set_xlabel('Value')
ax.set_ylabel('Density')
ax.set_title('Normalized Histograms of Price and Odometer (KM)')
ax.legend();
```

## Box Plots

```python
# Create a box plot for a given numerical data set.
# Generate multiple box plots in a single figure to compare different data sets.
# Customize the box plots with titles, labels, and change the colors.
```

```python
# Create a box plot for a given numerical data set.
np.random.seed(42)
data = np.random.randn(1000)
plt.boxplot(data);
```

```python
# Generate multiple box plots in a single figure to compare different data sets.
np.random.seed(42)
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 2, 100)
data3 = np.random.normal(-1, 1.5, 100)

fig, ax = plt.subplots(figsize=(10, 6))
ax.boxplot([data1, data2, data3]);
```

```python
# Customize the box plots with titles, labels, and change the colors.
fig, ax = plt.subplots(figsize=(10, 6))
box = ax.boxplot([data1, data2, data3], patch_artist=True);

ax.set(
    title='Box Plots of Data Sets',
    xlabel='Data Sets',
    ylabel='Values',
    xticklabels=['Data 1', 'Data 2', 'Data 3']
)

colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
```
