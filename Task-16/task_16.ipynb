```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## Basic Plots:

```python
# Create a simple line plot.
plt.plot([1, 2, 3, 4]);
```

```python
# Plot multiple lines on the same graph.

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 4, 6, 8]

fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2);
```

```python
# Add markers to a line plot.
fig, ax = plt.subplots()
ax.plot(x, y1, marker='o');
```

```python
plt.style.available
```

```python
plt.style.use('seaborn')
```

```python
# Customize line styles and colors.
fig, ax = plt.subplots()
ax.plot(x, y1, marker='o', color='red', linestyle='--');
```

## Scatter Plots:

```python
# Create a basic scatter plot.
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x, y);
```

```python
# Customize marker styles, sizes, and colors.
plt.scatter(x, y, marker='x', s=10, c='red');
```

```python
# Plot multiple scatter plots on the same graph.
fig, ax = plt.subplots()
ax.scatter(x, y, marker='o', s=10, c='red');
ax.scatter(y, x, marker='x', s=10, c='blue');
```

```python
# Add labels and titles to the scatter plot.
fig, ax = plt.subplots()
ax.scatter(x, y, marker='o', s=10, c='red');
ax.set(xlabel='X-axis', ylabel='Y-axis', title='Scatter Plot');
```

## Bar Charts

```python
### Bar Charts:
# Create a vertical bar chart.
# Create a horizontal bar chart.
# Customize bar colors and add edge colors.
```

```python
my_dict = {'A': 10, 'B': 40, 'C': 20}
fig, ax = plt.subplots()
ax.bar(my_dict.keys(), my_dict.values());
```

```python
# Create a horizontal bar chart.
fig, ax = plt.subplots()
ax.barh(list(my_dict.keys()), list(my_dict.values()));
```

```python
# Customize bar colors and add edge colors.
fig, ax = plt.subplots()
ax.bar(my_dict.keys(), my_dict.values(), color='red', edgecolor='black');
```

## Pie charts

```python
# Create a basic pie chart.
fig, ax = plt.subplots()
ax.pie(my_dict.values(), labels=my_dict.keys());
```

```python
# Customize slice colors and add a legend.
fig, ax = plt.subplots()
ax.pie(my_dict.values(), labels=my_dict.keys(), colors=['red', 'green', 'blue']);
ax.legend();
```

```python
# Explode a slice of the pie chart.
fig, ax = plt.subplots()
ax.pie(my_dict.values(), labels=my_dict.keys(), colors=['red', 'green', 'blue'], explode=[0.1, 0, 0]);
```

```python
# Add percentage labels to slices.
fig, ax = plt.subplots()
ax.pie(my_dict.values(), labels=my_dict.keys(), colors=['red', 'green', 'blue'], explode=[0.1, 0, 0], autopct='%1.1f%%');
```
