```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor, Lasso, ElasticNet, Ridge
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor, BaggingRegressor, StackingRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Load the dataset
housing = fetch_california_housing()
X = pd.DataFrame(housing.data, columns=housing.feature_names)
Y = pd.Series(housing.target)
```

```python
# Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Function to print and plot results
def print_and_plot_results(model, model_name):
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
    print(f'{model_name} RMSE: {rmse}')
    plt.scatter(Y_test, Y_pred, color='blue')
    plt.xlabel('Actual Prices')
    plt.ylabel('Predicted Prices')
    plt.title(f'{model_name} Results')
    plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color='red', linewidth=2)
    plt.show()
```

```python
# SGD Regressor
sgd_regressor = SGDRegressor(max_iter=1000, tol=1e-3, random_state=42)
print_and_plot_results(sgd_regressor, 'SGD Regressor')
```

```python
# Elastic Net Regression
elastic_net_regressor = ElasticNet(alpha=0.1, l1_ratio=0.5, random_state=42)
print_and_plot_results(elastic_net_regressor, 'Elastic Net Regression')
```

```python
# SVR (RBF Kernel)
svr_rbf = SVR(kernel='rbf', C=1.0, epsilon=0.1)
print_and_plot_results(svr_rbf, 'SVR (RBF Kernel)')
```

```python
# Ridge Regression
ridge_regressor = Ridge(alpha=1.0, random_state=42)
print_and_plot_results(ridge_regressor, 'Ridge Regression')
```

```python
# Random Forest Regressor
random_forest_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
print_and_plot_results(random_forest_regressor, 'Random Forest Regressor')
```

```python
# Ensemble Regressors
# Gradient Boosting Regressor
gbr = GradientBoostingRegressor(n_estimators=100, random_state=42)
print_and_plot_results(gbr, 'Gradient Boosting Regressor')
```

```python
# AdaBoost Regressor
abr = AdaBoostRegressor(n_estimators=100, random_state=42)
print_and_plot_results(abr, 'AdaBoost Regressor')
```

```python

# Bagging Regressor
br = BaggingRegressor(n_estimators=100, random_state=42)
print_and_plot_results(br, 'Bagging Regressor')
```
