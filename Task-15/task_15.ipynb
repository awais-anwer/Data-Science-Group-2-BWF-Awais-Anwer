```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('/content/df1.csv')
df2 = pd.read_csv('/content/df2.csv')
```

```python
# Merge two DataFrames on a single key.
df_merged = pd.merge(df1, df2, on='Key')
df_merged.head()
```

```python
# Merge two DataFrames on multiple keys.
```

```python
df1_multi = pd.DataFrame({
    'Key1': ['A', 'B', 'C', 'D'],
    'Key2': ['W', 'X', 'Y', 'Z'],
    'Value1': np.random.randn(4)
})

df2_multi = pd.DataFrame({
    'Key1': ['B', 'D', 'E', 'F'],
    'Key2': ['X', 'Z', 'W', 'Y'],
    'Value2': np.random.randn(4)
})

df_merged_multi = pd.merge(df1_multi, df2_multi, on=['Key1', 'Key2'])
df_merged_multi
```

```python
# Perform an outer join, inner join, left join, and right join.
outer_join = pd.merge(df1, df2, on='Key', how='outer')
inner_join = pd.merge(df1, df2, on='Key', how='inner')
left_join = pd.merge(df1, df2, on='Key', how='left')
right_join = pd.merge(df1, df2, on='Key', how='right')
```

```python
# Concatenate two DataFrames along rows.
```

```python
df3 = pd.read_csv('/content/df3.csv')
df3.head()
```

```python
concatenated_rows = pd.concat([df1, df3], axis=0)
concatenated_rows
```

```python
# Concatenate two DataFrames along columns.
concatenated_columns = pd.concat([df1, df2], axis=1)
concatenated_columns
```

```python
# Concatenate a list of DataFrames.
df_list = [df1, df2, df3]
concatenated_list = pd.concat(df_list, axis=0)
concatenated_list
```

```python
# Reshape data using the melt function to go from wide to long format.
```

```python
df_wide = pd.read_csv('/content/df_wide.csv')
df_wide
```

```python
# Reshape data using the melt function to go from wide to long format
df_long = pd.melt(df_wide, id_vars=['ID', 'A'], value_vars=['B', 'C'])
df_long
```

```python
# Create a pivot table to summarize data.
pivot_table = df_wide.pivot_table(index='A', values=['B', 'C'], aggfunc='sum')
pivot_table
```

```python
df_group = pd.read_csv('/content/df_group.csv')
df_group
```

```python
# Group data by one or more columns and perform aggregation functions (e.g., sum, mean, count).
df_group.groupby('Category').mean()
```

```python
# Apply multiple aggregation functions to grouped data.
df_wide.groupby('A').agg({'B': 'sum', 'C': ['mean', 'count']})
```

```python
# Use the groupby function to group data and apply custom functions.
df_wide.groupby('A').apply(lambda x: x.sum())
```
