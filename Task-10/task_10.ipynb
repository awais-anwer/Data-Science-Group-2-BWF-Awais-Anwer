```python
import numpy as np
```

```python
a1 = np.arange(10)
a1
```

```python
a2 = np.random.randint(1, 20, size=(3, 3))
a2
```

```python
a3 = np.ones(((2, 3, 4)))
a3
```

```python
# Add two 1D arrays element-wise.
arr1 = np.array([1, 2, 1])
arr2 = np.array([4, 2, 4])
arr1 + arr2
```

```python
# Multiply two 2D arrays element-wise.
arr1 = np.array([[1, 2],
                 [3, 4]])
arr2 = np.array([[5, 6],
                 [7, 8]])
print(arr1 * arr2)
```

```python
# Calculate the dot product of two matrices.
np.dot(arr1, arr2)
```

```python
# Calculate the mean, median, and standard deviation of a 1D array.
arr1 = np.array([2, 4, 5, 10, 15, 20, 21, 24, 30, 35])
print(np.mean(arr1))
print(np.median(arr1))
print(np.std(arr1))
```

```python
# Find the maximum and minimum values in a 2D array.
np.random.seed(42)
arr = np.random.randint(100, size=(3, 4))
print(arr)
print(np.max(arr))
print(np.min(arr))
```

```python
# Generate an array of 1000 random numbers from a normal distribution with a mean of 0 and a standard deviation of 1.
arr = np.random.normal(0, 1, 1000)
print(arr)
```

```python
# Create a 2D array of shapes (5, 5) with random integers from a uniform distribution between 10 and 50.

arr = np.random.randint(10, 50, size=(5, 5))
print(arr)
```

```python
# Calculate the cumulative sum of a 1D array.
arr = np.random.randint(10, size=15)
print(arr)
print(np.cumsum(arr))
```

```python
# Compute the correlation coefficient matrix of a 2D array.
arr = np.random.randint(0, 10, size=(5, 5))
print("matrix:")
print(arr)
print("correlation coefficient matrix:")
print(np.corrcoef(arr))
```

```python
# Simulate rolling a six-sided die 1000 times and count the frequency of each outcome.
rolling = np.random.randint(1, 7, 1000)
print(np.unique(rolling, return_counts=True))
```
