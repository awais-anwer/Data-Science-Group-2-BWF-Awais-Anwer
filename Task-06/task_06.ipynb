### Task 6 of Bytewise Fellowship. Exception handling, File Handling.

```python
filename = 'student_management_file.txt'

try:
    file = open(filename, 'r')
    # file exists no need to create
    file.close()
except FileNotFoundError:
    # file does not exist. creating the file
    file = open(filename, 'x')
    file.close()

# method to check unique id
def is_unique_id(student_id):
    with open(filename, 'r') as file:
        for line in file:
            existing_id = line.split(',')[0]
            if existing_id == student_id:
                return False
    return True

while True:
  print("--------------------Student Management System-------------------------")
  print("Choose one option:")
  print("1. Enter new student data")
  print("2. Search student data")
  print("3. Delete student data")
  print("4. View all student data")
  print("5. Exit")
  print("----------------------------------------------------------------------")

  choice = int(input("Enter your choice: "))

## option 1: Enter new student data
  if choice == 1:
    id = input("Enter student id:")
    if not is_unique_id(id):
      print("this Student id already exists. try again with a new id")
      continue
    name = input("Enter student name: ")
    age = int(input("Enter student age: "))
    grade = input("Enter student grade: ")
    file = open(filename, 'a')
    file.write(f"{id},{name},{age},{grade}\n")
    file.close()
    print("Student data added successfully.")

## option 2: Search a student's data
  elif choice == 2:
    search_id = input("Enter student id to search: ")
    found = False
    with open(filename, 'r') as file:
      for line in file:
        existing_id = line.split(',')[0]
        if existing_id == search_id:
          found = True
          print(f"Student name: {line.split(',')[1]}")
          print(f"Student age: {line.split(',')[2]}")
          print(f"Student grade: {line.split(',')[3]}")
          break
    if not found:
      print("Student not found.")

## option 3: delete a student's data
  elif choice == 3:
    delete_id = input("Enter student id to delete: ")
    found = False
    lines = []

    with open(filename, 'r') as file:
      for line in file:
        existing_id = line.split(',')[0]
        if existing_id != delete_id:
          lines.append(line)
        else:
          found = True

    with open(filename, 'w') as file:
      file.writelines(lines)

    if found == True:
      print("Student data deleted successfully.")
    else:
      print("Student not found.")

## option 4: view all student data
  elif choice == 4:
    with open(filename, 'r') as file:
      print(file.read())

## option 5: end program
  elif choice == 5:
    break

## no correct option chosen
  else:
    print("wrong input, try again")
```
