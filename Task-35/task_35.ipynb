```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('cat.jpg')
cv2_imshow(img)
cv2.destroyAllWindows()
```

```python
# Load an image in grayscale mode
img_gray = cv2.imread('cat.jpg', 0)
cv2_imshow(img_gray)
cv2.destroyAllWindows()
```

### 1. Enhancing the image

```python
enhanced_img = cv2.equalizeHist(img_gray)
cv2_imshow(enhanced_img)
cv2.destroyAllWindows()
```

### 2. Sharpening the image

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])

sharpened_img = cv2.filter2D(img, -1, kernel)
cv2_imshow(sharpened_img)
cv2.destroyAllWindows()
```

### 3. Blurring

```python
blurred_img = cv2.GaussianBlur(img, (15, 15), 0)
cv2_imshow(blurred_img)
cv2.destroyAllWindows()
```

### 4. Resizing

```python
resized_img = cv2.resize(img, (200, 200))
cv2_imshow(resized_img)
cv2.destroyAllWindows()
```

### 5. Rotation

```python
height, width = img.shape[:2]

rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), -45, 1)
rotated_img = cv2.warpAffine(img, rotation_matrix, (width, height))
cv2_imshow(rotated_img)
cv2.destroyAllWindows()
```

### 6. Flipping

```python
flipped_img = cv2.flip(img, 1)  # 1 = horizontal flip
cv2_imshow(flipped_img)
cv2.destroyAllWindows()
```
